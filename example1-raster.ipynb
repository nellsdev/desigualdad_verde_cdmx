```python
import rioxarray as rxr
import rasterio
import matplotlib.pyplot as plt
import numpy as np

print("✅ Librerías cargadas correctamente")
```

```python
# Usar un dataset de ejemplo de rasterio
from rasterio.plot import show

# O alternativamente, descargar un DEM pequeño:
import urllib.request
url = "https://github.com/rasterio/rasterio/raw/main/tests/data/RGB.byte.tif"
urllib.request.urlretrieve(url, "ejemplo.tif")
print("✅ Raster de ejemplo descargado")
```

```python
# Abrir con rioxarray (más amigable que rasterio puro)
raster = rxr.open_rasterio("ejemplo.tif")

# Ver metadatos básicos
print("📊 Metadatos del raster:")
print(f"  - Dimensiones: {raster.shape}")  # (bandas, filas, columnas)
print(f"  - CRS: {raster.rio.crs}")
print(f"  - Resolución: {raster.rio.resolution()}")
print(f"  - Extensión: {raster.rio.bounds()}")
print(f"  - Número de bandas: {raster.rio.count}")
print(f"  - Tipo de dato: {raster.dtype}")
```

```python
# Plotear la primera banda
fig, ax = plt.subplots(figsize=(10, 8))
raster[0].plot(ax=ax, cmap='viridis', add_colorbar=True)
ax.set_title("Banda 1 del raster de ejemplo")
plt.show()
```

```python
# Convertir a array NumPy (¡esto ya lo conoces!)
array_raster = raster.values[0]  # primera banda
print(f"📊 Estadísticas del array:")
print(f"  - Min: {np.nanmin(array_raster):.2f}")
print(f"  - Max: {np.nanmax(array_raster):.2f}")
print(f"  - Media: {np.nanmean(array_raster):.2f}")
print(f"  - Hay NaN?: {np.any(np.isnan(array_raster))}")
```
